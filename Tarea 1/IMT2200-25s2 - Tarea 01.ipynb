### 3.1 DataFrame unificado (1 punto)

Vamos a cargar en un DataFrame los datos de juegos guardados en el documento `bgg_GameItem.csv`. Al inspeccionar nuestros datos, podemos notar que los valores de ciertas comunas vienen "codificados" con ID. Para comenzar, vamos a juntar la información de los distintos archivos descargados.

**a) (0.8 pts)** Genere un DataFrame único con toda la información de cada juego, incluyendo: nombres de artistas y diseñadores involucrados, mecánicas, categorías, tipo, familia y editorial. Guarde este DataFrame en un nuevo archivo CSV.

```python
# Cargar todos los archivos CSV
data_category = pd.read_csv('archive/bgg_Category.csv')
data_game_family = pd.read_csv('archive/bgg_GameFamily.csv')
data_game_item = pd.read_csv('archive/bgg_GameItem.csv')
data_game_type = pd.read_csv('archive/bgg_GameType.csv')
data_mechanic = pd.read_csv('archive/bgg_Mechanic.csv')
data_person = pd.read_csv('archive/bgg_Person.csv')
data_publisher = pd.read_csv('archive/bgg_Publisher.csv')

# Revisar los tipos de datos
print(data_game_item.info())
```

```python
# Renombrar columnas de 'name' para evitar duplicados o cosas asi
data_category.rename(columns={'name': 'category_name'}, inplace=True)
data_game_family.rename(columns={'name': 'family_name'}, inplace=True)
data_game_type.rename(columns={'name': 'game_type_name'}, inplace=True)
data_mechanic.rename(columns={'name': 'mechanic_name'}, inplace=True)
data_person.rename(columns={'name': 'person_name'}, inplace=True)
data_publisher.rename(columns={'name': 'publisher_name'}, inplace=True)

# Corregir el tipo de dato de la columna 'game_type' antes de la fusión
data_game_item['game_type'] = pd.to_numeric(data_game_item['game_type'], errors='coerce')
```

```python
# Crear el DataFrame base y unir el tipo de juego
df_final = pd.merge(data_game_item, data_game_type, left_on='game_type', right_on='bgg_id', how='left')
df_final.drop(columns=['game_type', 'bgg_id_y'], inplace=True)
df_final.rename(columns={'bgg_id_x': 'bgg_id', 'name': 'game_name'}, inplace=True)

# Procesar y consolidar cada columna por separado
# Codigo sugerido por IA. Prompt: "Como puedo hacer el proceso de merge mas eficiente?"
def process_and_merge(df_base, df_lookup, id_col, name_col, final_col_name):
    temp_df = df_base[['bgg_id', id_col]].copy()
    temp_df[id_col] = temp_df[id_col].fillna("").astype(str).str.split(",")
    temp_df = temp_df.explode(id_col)
    temp_df[id_col] = pd.to_numeric(temp_df[id_col], errors='coerce')
    temp_df = temp_df.merge(df_lookup, left_on=id_col, right_on='bgg_id', how='left')
    
    # ** Línea corregida: Renombrar 'bgg_id_x' a 'bgg_id' después de la fusión **
    temp_df.rename(columns={'bgg_id_x': 'bgg_id'}, inplace=True)
    
    grouped_df = temp_df.groupby('bgg_id').agg(
        **{final_col_name: (name_col, lambda x: ', '.join(x.dropna().unique()))}
    ).reset_index()

    return pd.merge(df_base, grouped_df, on='bgg_id', how='left')

# Fin codigo sugerido por IA
```

```python
df_final = process_and_merge(df_final, data_person, 'designer', 'person_name', 'Designer')
df_final = process_and_merge(df_final, data_person, 'artist', 'person_name', 'Artist')
df_final = process_and_merge(df_final, data_publisher, 'publisher', 'publisher_name', 'Publisher')
```

```python

df_final = process_and_merge(df_final, data_category, 'category', 'category_name', 'Category')
df_final = process_and_merge(df_final, data_mechanic, 'mechanic', 'mechanic_name', 'Mechanic')
df_final = process_and_merge(df_final, data_game_family, 'family', 'family_name', 'Game_Family')
```

```python

# Limpiar las columnas de IDs originales
df_final.drop(columns=['designer', 'artist', 'publisher', 'category', 'mechanic', 'family'], inplace=True)

# Guardar el resultado en un archivo CSV
df_final.to_csv('merge_games.csv', index=False)

print("\n¡El DataFrame consolidado se ha creado y guardado como 'consolidated_games_fast.csv' exitosamente! ✅")
```

**a) (0.3 pts)** Inspecciones y filtre los datos que tengan valores válidos para: tiempo mínimo de juego, tiempo máximo de juego y complejidad. En el caso de que existan "outliers", puede descartarlos, cosiderando un rango razonable para las variables anteriores. Justifique su desición.

 *Eliminamos aquellos que tienen **valores superiores a los 1000**, ya que probablemente son errores de medicion. Tambien quitamos en caso de que exista algun juego con tiempo 0.*

```python
data_frame_by_time = data_frame_by_year.dropna(subset = ["complexity", "min_time", "max_time"])
data_frame_by_time = data_frame_by_time[(data_frame_by_time['min_time'] < 1000) & (data_frame_by_time['max_time'] < 1000)]
data_frame_by_time = data_frame_by_time[(data_frame_by_time['min_time'] >= 1) & (data_frame_by_time['max_time'] >= 1)]

filter_data_frame_by_time = data_frame_by_time[["game_name","year", "complexity", "min_time", "max_time"]].sort_values(by = 'min_time', ascending = False)

filter_data_frame_by_time
```

**b) (0.2 pts)** Veremos ahora si ha cambiado la duración promedio de los juegos de mesa en nuestro rango de años seleccionado. Primero, cree una nueva columna `avg_time` en el dataset con la duración promedio supuesta para cada juego.

```python

filter_data_frame_by_time["avg_time"] = (data_frame_by_time["min_time"] + data_frame_by_time["max_time"]) / 2
filter_data_frame_by_time = filter_data_frame_by_time[["game_name", "year", "complexity", "min_time", "max_time", "avg_time"]].sort_values(by='avg_time', ascending=False)
filter_data_frame_by_time
```

**c) (0.5 pts)** Grafique la duración promedio de juegos al año entre 1980 y 2025. ¿Existe alguna tendencia? Asegúrese de colocar las unidades correspondientes en sus ejes si es necesario. Comente sus resultados.

```python
avg_time_per_year = filter_data_frame_by_time.groupby("year")["avg_time"].mean()
plt.figure(figsize=(12, 6))
plt.bar(avg_time_per_year.index, avg_time_per_year.values, color='blue')
plt.title('Promedio de tiempo de juego por año')
plt.xlabel('Año')
plt.ylabel('Tiempo promedio (minutos)')


# Codigo sugerido por IA. Promt: "como puedo generar una linea de tendencia?"
x = avg_time_per_year.index.values
y = avg_time_per_year.values
z = np.polyfit(x, y, 1)
a, b = z
p = np.poly1d(z)
y_trend = p(x)
plt.plot(x, y_trend, color='red', linestyle='-', label='Línea de Tendencia')
# Fin codigo por IA

```
